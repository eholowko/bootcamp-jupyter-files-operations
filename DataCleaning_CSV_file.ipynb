```python
def split_col(df,col_index):
    col_i_name = df.columns[col_index]
    #print(col_i_name)
    res = re.match('[Z](?P<name>[\d]+)\((?P<l>[\d])[\+]*(?P<d>[\d]*)',col_i_name).groupdict()
    
    L_i = "L" + res["name"] + "(" + res["l"] + ")"
    #print(L_i)

    col_splitted = df.iloc[:,col_index].str.split('+',expand=True)
    
    if len(res["d"]) != 0:
        D_i = "D" + res["name"] + "(" + res["d"] + ")"
        #print(D_i)
        df[L_i] = col_splitted[0]
        df[D_i] = col_splitted[1]
    else:
        df[L_i] = col_splitted[0]        
    return df
```

## Oceny końcowe wg schematu: 
** [0%,50%) -> 2 
[51%,60%) -> 3 
[61%,70%) -> 3.5 
[71%,80%) -> 4 
 [81%,90%) -> 4.5 
 [91%,Inf] -> 5**

```python
sum_max = 0 #maksymalna suma punktów możliwa do uzyskania
for col in data2.iloc[:,3:data2.shape[1]].columns:
    sum_max += float(re.findall("\(([\d])\)",col)[0])
sum_max
```

```python
data2["suma"] = data2.sum(axis=1)
```

```python
data2["procenty"] = data2["suma"]/sum_max * 100
```

```python
data2
```

```python
data2["ocena"] = pd.cut(data2["procenty"],[0,50,60,70,80,90,100],labels=[2.0,3.0,3.5,4.0,4.5,5.0])
```
