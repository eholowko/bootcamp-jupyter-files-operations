## Funkcja zwracająca dla danej szkoły jakich obrażeń było najwięcej

```python
statistic = data[data["Typ"] == "Przedszkole"].iloc[:,2:5]
```

```python
statistic
```

```python
obj = statistic.idxmax(axis=1)
```

```python
obj[0]
```

```python
staistic[obj[0]][0]
```

```python
def obrazenia(typ):
    statistic = data[data["Typ"] == typ].iloc[:,2:5]
    obj = statistic.idxmax(axis=1)[0]
    return (obj,staistic[obj][0])
```

```python
obrazenia("Przedszkole")
```
